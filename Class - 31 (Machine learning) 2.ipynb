# What is underfitting and overfitting?

# Regularization

# L1,L2 :: Regularization

L1 : Lasso Regression

L2 : Rigde Regression

```python
from sklearn.linear_model import Lasso,Ridge
```

```python
# will remove the cofficent to zero (those features are not informative).

# alpha values could be .....> .0001,.001,.01,.1,1,10------>higher values reduce all the cofficients towards 0 and impact output

# defult value of alpha -1.0

ls = Lasso(alpha=0.0001)
ls.fit(x_train,y_train)
ls.score(x_train,y_train)
```

```python
ls.coef_
```

```python
#ridge regression

# try to minimize the cofficient variance

rd = Ridge(alpha=0.0001)
rd.fit(x_train,y_train)
rd.score(x_train,y_train)
```

```python
rd.coef_
```

```python
# ElasticNet is a combination of both Lasso and Ridge

from sklearn.linear_model import ElasticNet
```

```python
enr = ElasticNet(alpha = 0.0001)
enr.fit(x_train,y_train)
enr.score(x_train,y_train)
```

```python
enr.coef_
```

```python
from sklearn.svm import SVR
```

```python
# SVR (support victor regression)


svr = SVR(kernel = 'linear')
svr.fit(x_train,y_train)
svr.score(x_train,y_train)
pred_y = (x_test)
print(svr.score(x_train,y_train))


svr = SVR(kernel = 'poly')
svr.fit(x_train,y_train)
svr.score(x_train,y_train)
pred_y = (x_test)
print(svr.score(x_train,y_train))

svr = SVR(kernel = 'rbf')
svr.fit(x_train,y_train)
svr.score(x_train,y_train)
pred_y = (x_test)
print(svr.score(x_train,y_train))
```

```python
from sklearn.svm import SVR
kernellist=['linear','poly','rbf']

for i in kernellist:
    sv = SVR(kernel=i)
    sv.fit(x_train,y_train)
    sv.score(x_train,y_train)
    print(sv.score(x_train,y_train))
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knr = KNeighborsRegressor()
knr.fit(x_train,y_train)

knr.score(x_train,y_train)
predknr = knr.predict(x_test)
print(predknr)

knr.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor()
dtr.fit(x_train,y_train)
dtrpred = dtr.predict(x_test)
print(dtrpred)

dtr.score(x_test,y_test)
```
